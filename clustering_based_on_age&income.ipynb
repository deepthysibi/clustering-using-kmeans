# full set iris data

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
#from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
idf=pd.read_csv("iris.csv")
idf
```

```python
df=idf.drop(["Species","Id"],axis=1)
df
```

```python
idf["Species"].value_counts()
```

```python
sns.pairplot(df)
```

```python
iris=idf.drop("Id",axis=1)
sns.set_style("whitegrid")
sns.pairplot(iris,hue="Species",size=3);
plt.show()
```

```python
ssei=[]
for k in range(1,20):
    kmi=KMeans(n_clusters=k)
    kmi.fit(df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
    ssei.append(kmi.inertia_)
```

```python
plt.plot(range(1,20),ssei)
plt.xlabel("k value")
plt.ylabel("sse")
```

```python
#consider k=3
kmi=KMeans(n_clusters=3)
ipred=kmi.fit_predict(df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
ipred
```

```python
df["cluster"]=ipred
df
```

```python
centers=kmi.cluster_centers_
centers
```

```python
ndf1=df[df['cluster']==0]
ndf2=df[df['cluster']==1]
ndf3=df[df['cluster']==2]
```

```python
ndf1
```

```python
plt.scatter(ndf1['SepalLengthCm'],ndf1["SepalWidthCm"],c="purple",label="Iris-versicolor")
plt.scatter(ndf2['SepalLengthCm'],ndf2["SepalWidthCm"],c="orange",label="Iris-setosa")
plt.scatter(ndf3['SepalLengthCm'],ndf3["SepalWidthCm"],c="green",label="Iris-virginica")
plt.xlabel("SepalLengthCm")
plt.ylabel("sepelwidth")
plt.scatter(centers[:,0],centers[:,1],marker="*",color="black",label="center")
plt.legend()
```

```python

fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
plt.scatter(ndf1['SepalLengthCm'],ndf1["SepalWidthCm"],c="purple",s=100,label="Iris-versicolor")
plt.scatter(ndf2['SepalLengthCm'],ndf2["SepalWidthCm"],c="orange",s=100,label="Iris-setosa")
plt.scatter(ndf3['SepalLengthCm'],ndf3["SepalWidthCm"],c="green",s=100,label="Iris-virginica")

#Plotting the centroids of the clusters
plt.scatter(centers[:,0],centers[:,1],marker="*",color="black",s=100,label="center")
plt.show()
```

# Mall segmentyation data from kaggle

```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
df=pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
x
```

```python
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"])
plt.xlabel("Income")
plt.ylabel("Spending score")
```

```python
sse=[]
for k in range(1,50):
    km=KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
plt.plot(range(1,50),sse)
plt.xlabel("K value")
plt.ylabel("sse")
```

```python
sse=[]
for k in range(1,11):
    km=KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
plt.plot(range(1,11),sse)
plt.xlabel("K value")
plt.ylabel("sse")    
```

```python
#elbow at 5 so take k=5
km=KMeans(n_clusters=5)
pred=km.fit_predict(x)
pred
```

```python
df["cluster"]=pred
df
```

```python
centroid=km.cluster_centers_
centroid
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[df["cluster"]==0]["Annual Income (k$)"],df[df["cluster"]==0]["Spending Score (1-100)"],color="red",label="C-1")
plt.scatter(df[df["cluster"]==1]["Annual Income (k$)"],df[df["cluster"]==1]["Spending Score (1-100)"],color="green",label="C-2")
plt.scatter(df[df["cluster"]==2]["Annual Income (k$)"],df[df["cluster"]==2]["Spending Score (1-100)"],color="purple",label="C-3")
plt.scatter(df[df["cluster"]==3]["Annual Income (k$)"],df[df["cluster"]==3]["Spending Score (1-100)"],color="yellow",label="C-4")
plt.scatter(df[df["cluster"]==4]["Annual Income (k$)"],df[df["cluster"]==4]["Spending Score (1-100)"],color="cyan",label="C-5")
plt.scatter(centroid[:,0],centroid[:,1],marker="*",color="black",label="center")
plt.legend()
plt.xlabel("Income")
plt.ylabel("spending score")
plt.title("Clustered diagram")
```
